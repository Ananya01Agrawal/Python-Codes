```python
# we need to do the comparative as well as combining study
import matplotlib.pyplot as plt
import numpy as np
year=np.array([2015,2016,2017,2018,2019,2020,2021])
tata=np.array([1500,3400,4100,4700,5000,4300,3200])
mt=np.array([3000,4200,5200,4800,7000,6200,6000])
merz=np.array([5000,4400,3000,2100,1800,1000,800])
audi=np.array([1000,1800,1200,1800,1700,1200,1500])
plt.subplot(2,2,1)
plt.barh(year,tata,label="Tata",height=.5)
plt.barh(year,mt,label="MT",height=.5,left=tata)
plt.barh(year,merz,label="merz",height=.5,left=tata+mt)
plt.barh(year,audi,label="audi",height=.5,left=tata+mt+merz)
plt.subplot(2,2,4)
plt.bar(year,tata,label="Tata",width=.1,color="red")
plt.bar(year+.1,mt,label="MT",width=.1,color="black")
plt.bar(year+.2,mt,label="merz",width=.1,color="green")
plt.bar(year+.3,mt,label="audi",width=.1,color="violet")
plt.xlabel("years")
plt.ylabel("unit sold")
plt.title("car sale record after 7 years")
plt.grid()
plt.legend()
plt.show()
```
